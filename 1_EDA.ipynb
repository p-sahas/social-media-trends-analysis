## Exploratory Data Analysis (EDA)

### 1. Setup and Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed data
try:
    df = pd.read_csv('data/processed/processed_data.csv')
    print(f"Dataset Shape: {df.shape}")
    display(df.head())
except FileNotFoundError:
    print("Processed data not found. Please run 0_data_preparation.ipynb first.")
```

### 2. Correlation Analysis

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### 3. Distribution of Key Features

```python
features = ['Engagement_Rate', 'Likes_Per_View', 'Shares_Per_View', 'Comments_Per_View']

for feature in features:
    plt.figure(figsize=(5, 3))
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

### 4. Engagement by Time

```python
if 'Month' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Month', y='Engagement_Rate', data=df)
    plt.title('Average Engagement Rate by Month')
    plt.show()

if 'DayOfWeek' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='DayOfWeek', y='Engagement_Rate', data=df)
    plt.title('Average Engagement Rate by Day of Week (0=Mon, 6=Sun)')
    plt.show()
```
